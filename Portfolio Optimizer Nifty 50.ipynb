### Read List of Nifty 50 stocks

```python
import pandas as pd
A = pd.read_csv("D:/Nifty 50 index list/ind_nifty50list.csv")
A
```

```python
A['yfinance ticker'] = A['Symbol']+'.NS'
```

```python
tickers = list(A['yfinance ticker'])
```

```python
import yfinance as yf
d = yf.download(tickers,period='ytd',interval='1d')
d.head()
```

```python
df= d['Adj Close']
df.head()
```

### Import Python Portfolio Optimisation Library

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
mu = expected_returns.mean_historical_return(df)
s = risk_models.sample_cov(df)
```

```python
ef = EfficientFrontier(mu,s)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose=True)
```

### Allocate shares for value of 10000 INR 

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

portfolio_val = 10000
latest_prices = get_latest_prices(df)
weights = cleaned_weights
da = DiscreteAllocation(weights, latest_prices,total_portfolio_value=portfolio_val)
allocation, leftover = da.lp_portfolio()
print('Discrete Allocation : ',allocation)
print('Funds Remaining     : ',round(leftover,2),'INR')
```

```python
import pandas as pd
al_key = allocation.keys()
al_values = allocation.values()
dfa = pd.DataFrame([al_key,al_values]).T
dfa.columns = ['yfinance ticker','Discrete Quantity']
```

```python
dfa
```

## Show how much Stocks to invest in which company

```python
inner_join = pd.merge(A, 
                      dfa, 
                      on ='yfinance ticker', 
                      how ='inner')
```

```python
inner_join.sort_values(by='Discrete Quantity',ascending=False)
```
